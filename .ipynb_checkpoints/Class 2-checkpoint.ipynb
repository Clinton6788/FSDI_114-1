# Linked Lists
In general, there are two varieties of linked lists.

1. Singly linked lists -- aka unidirectional lists
2. Doubly linked lists -- aka bidirectional lists.

```python
class SinglyLinkedList():
    class __Node:
        def __init__(self, datum):
            self.data = datum
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.len = 0

    def append(self,datum):
        new_node = self.__Node(datum)
        if not self.tail:
            self.head = new_node
            self.tail = new_node
        else:
            old_tail = self.tail
            old_tail.next= new_node
            self.tail = new_node
        self.len += 1
            
        
    def insert(self, index, datum):
        if not isinstance(index, int):
            raise TypeError("Index must be an integer")
    
        new_node = self.__Node(datum)

        # Insert at head
        if index <= 0:
            new_node.next = self.head
            self.head = new_node
            if not self.tail: 
                self.tail = new_node
            self.len += 1
            return
    
        # Index > len
        if index >= self.len:
            self.append(datum)
            return
    
        # Inserting inside list @ specific pos
        current = self.head
        count = 0
    
        while current:
            if count == index - 1:  
                new_node.next = current.next
                current.next = new_node
                
                if new_node.next is None: # Check for tail update
                    self.tail = new_node
                self.len += 1
                return
            current = current.next
            count += 1

        # insert adds BEFORE target index
        # insert before head, between
        # defaults to append when index exceeds length of list
        
        
    def remove(self,datum):
        current = self.head
        found = False
        prev = None
        while current and not found:
            if current.data == datum:
                found = True
            else:
                prev = current
                current = current.next
        if found:
            if not prev:
                self.head = self.head.next
            else:
                prev.next = current.next
        else:
            raise ValueError("Target value not found")
            
        self.len -= 1
        pass
        
    def is_empty(self):
        return True if self.len <= 0 else False
        
    def index(self, datum):
        current = self.head
        count = 0
        while current:
            if current.data == datum:
                return count
            if count == self.len:
                raise IndexError("Data not found")
            current = current.next
            count += 1
            
        # returns first pos where datum exists in list
        # raises IndexError(data not found)
        
        
    def __len__(self):
        return self.len
        
    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out
    
```

```python
sll = SinglyLinkedList()
print(sll)
print(sll.is_empty())
sll.insert(5,"z")
sll.append(1)
print(sll)
sll.append(2)
print(sll)
sll.insert(20,"y")
print(sll)
sll.insert(1,"A")
print(sll)
print(sll.is_empty())
```
