```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, e):
        self.items.append(e)

    def pop(self):
        return self.items.pop()

    def is_empty(self):
        return self.items == []

    def peek(self):
        if not self.is_empty():
            return self.items[len(self.items)-1]
        else:
            return print("list is empty")

    def size(self):
        return len(self.items)
```

```python
s = Stack()

for n in range(6):
    s.push(n)

result= s.size()
print(result)
print(s.peek())

for _ in range(s.size()):
    print(s.pop())
print(s.size())
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, e):
        self.items.append(e)

    def dequeue(self):
        return self.items.pop(0)

    def is_empty(self):
        return self.items == []

    def peek(self):
        if not self.is_empty():
            return self.items[0]
        else:
            return None
            
    def size(self):
        return len(self.items)
    
```

```python
x=[i for i in range(11)]

def delay_items(l):
    q = Queue()
    for n in l:
        q.enqueue(n)
    while not q.is_empty():
        print(q.dequeue())
delay_items(x)
```
